# Dictionary {}

```python
d = {
    "name" : "Pinky",
    "age" : 23,    
    "marks" : [90, 78, 67.5, 100, 44],    
}

print(d)
print(d["name"])
print(d["marks"])
print(d["marks"][3])

d["age"] = 50

print(d)

d["city"] = "Jamshedpur"

print(d)

d["city"] = "Delhi"

print(d)
```
