# Input

```python
salary = int(input("Please Enter your Salary"))

if salary <= 10000:
    bonus = 2
    
elif salary <= 20000:
    bonus = 3
    
elif salary <= 30000:
    bonus = 5
    
elif salary <= 40000:
    bonus = 10
    
else:
    bonus = 15
    
print("Base Salary: ", salary)
bonus = salary * bonus/100

print("Bonus: ", bonus)
print("Final Salary: ", salary + bonus)
```

```python
lucky_number = 101010
lottery = 0

while lottery != lucky_number:
    lottery = int(input("What is your lottery number?"))
    print(lottery)
else:
    print("Congratulationssss!!!!")    
```

```python
lucky_number = 101010
lottery = 0

while lottery != lucky_number:
    lottery = int(input("What is your lottery number?"))
    if lottery == lucky_number:
        break
    print(lottery, "Better Luck next time")

print("Congratulationssss!!!!")    
```

# Prime Number

```python
n = int(input("Enter Number: "))
ans = True

for i in range(2, n):
    if n%i == 0:
        ans = False
        
if ans == True:
    print(n, "is Prime")
else:
    print(n, "is Composite")
    
```
